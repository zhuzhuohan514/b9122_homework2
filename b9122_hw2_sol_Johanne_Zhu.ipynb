```python
from bs4 import BeautifulSoup
import urllib.request
import requests
import re
#from urllib.request import Request
```

# Crawl pages whose seed url is the press releases page of the Federal Reserve System

```python
seed_url = "https://www.federalreserve.gov/newsevents/pressreleases.htm"
seed_url_comp = "https://www.federalreserve.gov/newsevents/pressreleases"
urls = [seed_url]
seen = [seed_url] 
opened = []
result = []
```

```python
maxNumUrl = 10
maxCovidUrl = 10
print("Starting with url="+str(urls))
while len(urls) > 0 and len(opened) < maxNumUrl:
    try:
        curr_url=urls.pop(0)
        print("num. of URLs in stack: %d " % len(urls))
        print("Trying to access= "+curr_url)
        req = urllib.request.Request(curr_url,headers={'User-Agent': 'Mozilla/5.0'})
        webpage = urllib.request.urlopen(req).read()
        opened.append(curr_url)

    except Exception as ex:
        print("Unable to access= "+curr_url)
        print(ex)
        continue 

    soup = BeautifulSoup(webpage)
    for tag in soup.find_all('a', href = True): 
        childUrl = tag['href'] 
        o_childurl = childUrl
        childUrl = urllib.parse.urljoin(seed_url, childUrl)
        #print("childurl=" + childUrl)
        if seed_url_comp in childUrl and childUrl not in seen:
            urls.append(childUrl)
            seen.append(childUrl)
            newpageText = requests.get(childUrl).text
            if len(re.compile("covid", re.IGNORECASE).findall(newpageText)) > 0:
                result.append(childUrl)
                print(childUrl)
                if len(result) > maxCovidUrl:
                    break

    if len(result) > maxCovidUrl:
        break
print("num. of URLs seen = %d, and scanned = %d" % (len(seen), len(opened)))
```

```python
print("Top 10 Federal Reserve System links containing keyword covid:")
for i in result:
    print(i)
```

# Crawl pages whose seed url is the press releases page of the Securities and Exchange Commission

```python
seed_url = "https://www.sec.gov/news/pressreleases"
seed_url_comp = "https://www.sec.gov/news/press-release/"
urls = [seed_url]
seen = [seed_url] 
opened = []
result = []
```

```python
maxNumUrl = 10
maxChargeUrl = 20
print("Starting with url="+str(urls))
while len(urls) > 0 and len(opened) < maxNumUrl:
    try:
        curr_url=urls.pop(0)
        print("num. of URLs in stack: %d " % len(urls))
        print("Trying to access= "+curr_url)
        req = urllib.request.Request(curr_url,headers={'User-Agent': 'Mozilla/5.0'})
        webpage = urllib.request.urlopen(req).read()
        opened.append(curr_url)

    except Exception as ex:
        print("Unable to access= "+curr_url)
        print(ex)
        continue 

    soup = BeautifulSoup(webpage)
    for tag in soup.find_all('a', href = True): 
        childUrl = tag['href'] 
        o_childurl = childUrl
        childUrl = urllib.parse.urljoin(seed_url, childUrl)
        #print("childurl=" + childUrl)
        if seed_url_comp in childUrl and childUrl not in seen:
            urls.append(childUrl)
            seen.append(childUrl)
            newpageText = requests.get(childUrl).text
            if len(re.compile("charges", re.IGNORECASE).findall(newpageText)) > 0:
                result.append((childUrl, newpageText))
                print(childUrl)
                if len(result) > maxChargeUrl:
                    break

    if len(result) > maxChargeUrl:
        break
print("num. of URLs seen = %d, and scanned = %d" % (len(seen), len(opened)))
```

```python
print("Top 10 Securities and Exchange Commission links & texts containing keyword charges:")
for i in result:
    print(i)
    print("----------------------------------------")
```
